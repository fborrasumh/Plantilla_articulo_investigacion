Una investigación de los Estados Unidos dictaminó que el error de interceptación en Dharan se debió a un error de software en el reloj del sistema. La batería Patriot de Dharan había estado activada durante 100 horas, tras las cuales el reloj se había retrasado un tercio de segundo, equivalente a un error de posición de 600 metros.
